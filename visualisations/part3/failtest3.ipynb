```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm 
import pylab as py 
import seaborn as sns


print("=== 15 shapiro test ===")
net15  = pd.read_csv("stat-test-15-5000.csv")
w15_sub,pvalue15_sub = stats.shapiro(net15.sublinear)
w15_spv,pvalue15_spv = stats.shapiro(net15.linear)
print("sub w:",w15_sub,"sub p_value",pvalue15_sub)
print("SPV w:",w15_spv,"SPV p_value",pvalue15_spv)

print("=== 15 probability test ===")
stats.probplot(net15.sublinear, plot=sns.mpl.pyplot)
py.ylim(0.98,1)
py.show() 

stats.probplot(net15.linear, plot=sns.mpl.pyplot)
py.ylim(0.98,1)
py.show() 

print("=== 15 t test ===")
T,pvalue = stats.ttest_ind(net15.sublinear,net15.linear)
print("T:",T,"pvalue",pvalue)
```
