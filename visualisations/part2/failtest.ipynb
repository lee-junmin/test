```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm 
import pylab as py 
import seaborn as sns


print("=== 15 shapiro test ===")
net15  = pd.read_csv("stat-test-15-5000.csv")
w15_sub,pvalue15_sub = stats.shapiro(net15.scv)
w15_spv,pvalue15_spv = stats.shapiro(net15.spv)
print("sub w:",w15_sub,"sub p_value",pvalue15_sub)
print("SPV w:",w15_spv,"SPV p_value",pvalue15_spv)
print(stats.describe(net15.scv))
print(stats.describe(net15.spv))



print("=== 15 probability test ===")
stats.probplot(net15.scv, plot=sns.mpl.pyplot)
#py.ylim(0.98,1)
py.xlim(-4.5,4.5)
py.show() 

stats.probplot(net15.spv, plot=sns.mpl.pyplot)
#py.ylim(0.98,1)
py.xlim(-4.5,4.5)
#py.xlim(-8,8)
py.show() 

print("=== 15 t test ===")
#T,pvalue = stats.ttest_ind(net15.scv,net15.spv)
U,pvalue = stats.mannwhitneyu(net15.scv,net15.spv)
print("U:",U,"pvalue",pvalue)
```
